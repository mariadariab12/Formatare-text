```python
import pandas as pd
import os

def format_table_from_excel(file_path: str) -> str:
    df = pd.read_excel(file_path, sheet_name=0)

    df = df.apply(lambda col: col.map(lambda x: str(x).replace('"', '').replace('\n', ' ') if isinstance(x, str) else x))
    df = df.fillna("&nbsp;")

    header = "||" + "||".join(df.columns) + "||\n"
    rows = ""

    for _, row in df.iterrows():
        rows += "|" + "|".join(str(cell) for cell in row) + "|\n"

    return header + rows


xlsx_files = [f for f in os.listdir('/content') if f.endswith('.xlsx')]

if not xlsx_files:
    print("Nu există fișiere .xlsx în /content. Încarcă unul cu files.upload().")
else:
    file_path = os.path.join('/content', xlsx_files[0])

print(format_table_from_excel(file_path))
```
